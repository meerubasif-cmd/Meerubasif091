```python
nums = [9, 2, 11, 7, 11, 3]

largest = float('-inf')
second_largest = float('-inf')

for num in nums:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print(second_largest) 
```

```python
a = [5, 6, 7, 8, 9, 10]
k = int(input("Enter index k to delete: "))
for i in range(k, len(a) - 1):
    a[i] = a[i + 1]
a.pop()
print(a)
```

```python
text = input("Enter a line of text: ")
words = text.split()
longest_word = ""
max_length = 0
for word in words:
    if len(word) > max_length:
        longest_word = word
        max_length = len(word)
print(f"Longest word: {longest_word}")
print(f"Length: {max_length}")
```

```python

sentence = input("Enter a sentence: ")
words = sentence.split()
result = " | ".join(words)
print(result)
```

```python
a = [3, 3, 2, 3, 1, 2, 4, 4]
unique_list = []
for x in a:
    if x not in unique_list:
        unique_list.append(x)
print(unique_list)
```

```python
a = [1, 1, 1, 2, 2, 3, 4, 4]
seen = []
for num in a:
    if num not in seen:
        count = a.count(num)
        print(f"Value: {num}, Frequency: {count}")
        seen.append(num)
```

```python
a = [8, 3, 5, 2, 9, 1]
n = len(a)
for i in range(n - 1):
    for j in range(n - 1 - i):
        if a[j] > a[j + 1]:
            a[j], a[j + 1] = a[j + 1], a[j]
print(a)
```
